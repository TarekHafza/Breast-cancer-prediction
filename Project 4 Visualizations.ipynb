```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('BC_ML_tbl.csv')
df.head()
```

```python
sns.violinplot(x='radius_mean', y='texture_mean', data=df, hue='Diagnosis', split=True)
```

```python
df['texture_mean_binned'] = pd.cut(df['texture_mean'], bins=3, labels=["Low", "Medium", "High"])
df['smoothness_mean_binned'] = pd.cut(df['smoothness_mean'], bins=3, labels=["Low", "Medium", "High"])
df_fg = sns.FacetGrid(df, col='texture_mean_binned', row='smoothness_mean_binned')
df_fg.map(plt.hist, 'Diagnosis', bins=10)
```
